```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

# Defining Model

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25,10))
num_centers = 4
X, c = make_blobs(n_samples=500, n_features=2, centers=num_centers, cluster_std=[[np.random.uniform(.1, 3), np.random.uniform(.1, 3)] for _ in range(num_centers)])
ax[0].scatter(*X.T, c=c, cmap='viridis')

# Getting initial locations.
kmeans = KMeans(n_clusters=num_centers).fit(X)
mu_k = kmeans.cluster_centers_

gmm = GMM(num_centers, initial_mus=mu_k, initial_covs=[np.identity(2)*0.1]*num_centers, verbose=False)
gmm.fit(X, iters=10)
gmm.display_gmm(X, ax[1])
```
